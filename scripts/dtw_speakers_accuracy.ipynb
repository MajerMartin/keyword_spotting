```python
import os
import numpy as np
import common as common
```

```python
current_dir = os.getcwd()
results_dir = os.path.join(current_dir, '..', 'results')
data_dir = os.path.join(current_dir, '..', 'data')
samples_dir = os.path.join(data_dir, 'segmented')
dm_dir = os.path.join(data_dir, 'distance_matrices', 'dtw')

print 'Current directory:\t', current_dir
print 'Results directory:\t', results_dir
print 'Data directory:\t\t', data_dir
print 'Samples directory:\t', samples_dir
print 'Dist. mat. directory:\t', dm_dir

references = common.load_references(samples_dir)
speakers = [str(x + 1) for x in xrange(6)]
```

```python
fts = ['ste_10_10', 'sti_10_10', 'stzcr_10_10', 'ste_sti_stzcr_10_10', 'log_fb_en_25_10_ham', 'log_fb_en_25_10_ham_deltas', 'mfcc_25_10_ham', 'mfcc_25_10_ham_deltas']
fts_norm = ['ste_10_10_norm', 'sti_10_10_norm', 'stzcr_10_10_norm', 'ste_sti_stzcr_10_10_norm', 'log_fb_en_25_10_ham_norm', 'log_fb_en_25_10_ham_deltas_norm', 'mfcc_25_10_ham_norm', 'mfcc_25_10_ham_deltas_norm']
fts_bottleneck = ['BN3bip_MM_8', 'BN3bip_MM_16', 'BN3bip_MM-SA_8', 'BN3bip_MM-SA_16', 'BN3bip_MM-SA_32']
```

## Single speaker

```python
features = fts + fts_norm + ['ste_sti_stzcr_10_10_norm_optim_single'] + fts_bottleneck
accuracy_data = {}

for feature in features:
    distance_matrix, ref_keys_all, test_keys_all = common.get_data(dm_dir, feature)
    
    feature_accuracies = []
    
    for speaker in speakers:
        ref_keys = common.get_speaker_keys(ref_keys_all, speaker)
        test_keys = common.get_speaker_keys(test_keys_all, speaker) 
        
        recognition = common.recognize(test_keys, ref_keys, distance_matrix)
        accuracy =  common.compute_accuracy(test_keys, recognition, references)
        feature_accuracies.append(accuracy)

    feature_accuracies.append(np.mean(feature_accuracies))
    accuracy_data[feature] = feature_accuracies

dtw_single = common.create_dataframe(accuracy_data, speakers + ['Mean'])
dtw_single    
```

```python
csv_name = os.path.join(results_dir, 'dtw_single_speaker.csv')

with open(csv_name, 'w') as fw:
    dtw_single.to_csv(fw)
```

## All speakers

```python
features = fts + fts_norm + ['ste_sti_stzcr_10_10_norm_optim_all'] + fts_bottleneck
accuracy_data = {}

for feature in features:
    distance_matrix, ref_keys_all, test_keys_all = common.get_data(dm_dir, feature)
    
    ref_keys = []
    test_keys = []  
    
    for speaker in speakers:
        ref_keys.extend(common.get_speaker_keys(ref_keys_all, speaker))
        test_keys.extend(common.get_speaker_keys(test_keys_all, speaker)) 
        
    recognition = common.recognize(test_keys, ref_keys, distance_matrix)
    accuracy_data[feature] = common.compute_accuracy(test_keys, recognition, references)

dtw_all = common.create_dataframe(accuracy_data, ['All speakers'])
dtw_all    
```

```python
csv_name = os.path.join(results_dir, 'dtw_all_speakers.csv')

with open(csv_name, 'w') as fw:
    dtw_all.to_csv(fw)
```

## All test data against one speaker

```python
csv_name = os.path.join(results_dir, 'dtw_all_test_per_speaker.csv')

with open(csv_name, 'w') as fw:
    dtw_all_test_per_speaker.to_csv(fw)
```
